```python
# Hands-On #1: Vectors, Dot Products & Norms in NumPy

import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

# --------------------------------------------------
# Step 1: Create vectors for three students
# Format: [Math, Science, English]
# --------------------------------------------------

student_A = np.array([85, 90, 88])
student_B = np.array([80, 92, 85])
student_C = np.array([60, 70, 65])

print("Student A Scores:", student_A)
print("Student B Scores:", student_B)
print("Student C Scores:", student_C)

# --------------------------------------------------
# Step 2: Compute L2 Norms (Overall Performance)
# --------------------------------------------------

norm_A = np.linalg.norm(student_A)
norm_B = np.linalg.norm(student_B)
norm_C = np.linalg.norm(student_C)

print("\nL2 Norms (Overall Performance)")
print("Student A:", round(norm_A, 2))
print("Student B:", round(norm_B, 2))
print("Student C:", round(norm_C, 2))

# --------------------------------------------------
# Step 3: Compute Cosine Similarity
# --------------------------------------------------

sim_AB = cosine_similarity([student_A], [student_B])[0][0]
sim_AC = cosine_similarity([student_A], [student_C])[0][0]
sim_BC = cosine_similarity([student_B], [student_C])[0][0]

print("\nCosine Similarities")
print("A vs B:", round(sim_AB, 4))
print("A vs C:", round(sim_AC, 4))
print("B vs C:", round(sim_BC, 4))

# --------------------------------------------------
# Step 4: Dot Products (Optional)
# --------------------------------------------------

dot_AB = np.dot(student_A, student_B)
dot_AC = np.dot(student_A, student_C)
dot_BC = np.dot(student_B, student_C)

print("\nDot Products")
print("A · B =", dot_AB)
print("A · C =", dot_AC)
print("B · C =", dot_BC)

# --------------------------------------------------
# Step 5: Interpretation
# --------------------------------------------------

print("\nInterpretation:")

if sim_AB > sim_AC and sim_AB > sim_BC:
    print("Students A and B have the most similar score patterns.")

if norm_A > norm_B and norm_A > norm_C:
    print("Student A has the highest overall performance.")
elif norm_B > norm_A and norm_B > norm_C:
    print("Student B has the highest overall performance.")
else:
    print("Student C has the highest overall performance.")
```

Create three vectors representing student exam scores in [Math, Science, English].

```python
import numpy as np

# Create vectors for three students' exam scores [Math, Science, English]
student1_scores = np.array([85, 90, 88])
student2_scores = np.array([78, 82, 91])
student3_scores = np.array([65, 70, 75])

print("Student 1 Scores:", student1_scores)
print("Student 2 Scores:", student2_scores)
print("Student 3 Scores:", student3_scores)
```

Compute L2 norms (overall performance), cosine similarity between pairs (similar patterns?).

```python
from sklearn.metrics.pairwise import cosine_similarity

# Compute L2 Norms (Overall Performance)
norm_student1 = np.linalg.norm(student1_scores)
norm_student2 = np.linalg.norm(student2_scores)
norm_student3 = np.linalg.norm(student3_scores)

print("\nL2 Norms (Overall Performance):")
print(f"Student 1: {norm_student1:.2f}")
print(f"Student 2: {norm_student2:.2f}")
print(f"Student 3: {norm_student3:.2f}")

# Compute Cosine Similarity between pairs
sim_s1s2 = cosine_similarity([student1_scores], [student2_scores])[0][0]
sim_s1s3 = cosine_similarity([student1_scores], [student3_scores])[0][0]
sim_s2s3 = cosine_similarity([student2_scores], [student3_scores])[0][0]

print("\nCosine Similarities:")
print(f"Student 1 vs Student 2: {sim_s1s2:.4f}")
print(f"Student 1 vs Student 3: {sim_s1s3:.4f}")
print(f"Student 2 vs Student 3: {sim_s2s3:.4f}")
```
