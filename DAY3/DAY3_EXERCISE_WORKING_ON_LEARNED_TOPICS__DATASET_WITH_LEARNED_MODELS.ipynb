**20: Covariance**

```python
covariance = day_df['temp'].cov(day_df['cnt'])

print("Covariance:", covariance)
```

**21: Probability of High Demand**

```python
high_demand_days = len(day_df[day_df['cnt'] > 5000])

total_days = len(day_df)

probability = high_demand_days / total_days

print("Probability:", probability)
```

**22: Conditional Probability**

Probability of rentals > 5000 given Summer

```python
summer = day_df[day_df['season'] == 2]

high_summer = len(
    summer[summer['cnt'] > 5000]
)

conditional_probability = high_summer / len(summer)

print(conditional_probability)
```
